```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('/content/Customer Call List.xlsx') #Sample customer list specifically for data cleaning
```

```python
df
```

```python
df.info() #Finding information about all the columns that are present
```

```python
df.drop(columns=['Not_Useful_Column'],inplace=True) #Dropping the columns that are not useful
```

```python
df
```

```python
#Reformating the Paying Customer column and the Do_Not_Contact column to "Y" and "N"
df["Paying Customer"].replace("Yes","Y",inplace=True)
df["Paying Customer"].replace("No","N",inplace=True)
```

```python
df["Do_Not_Contact"].replace("Yes","Y",inplace=True)
df["Do_Not_Contact"].replace("No","N",inplace=True)
```

```python
df
```

```python
df.drop(index=df[df["Do_Not_Contact"]=="Y"].index,inplace=True) #Removing rows where do not contact is "Y", it is not useful.
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Last_Name"]=df["Last_Name"].str.strip("123/._") #stripping all characters from Last name.
```

```python
df["Phone_Number"]=df["Phone_Number"].str.strip("-|/") #No effect.Going to replace instead.
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')  #Replaced the characters with "Nothing"
```

```python
#Formatting the Phone number to 123-456-7890
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3]+"-"+x[3:6]+"-"+x[6:10])
```

```python
df
```

```python
df.replace('Na',np.nan,inplace=True)
df.replace('NaN',np.nan,inplace=True)
df.replace('N/a',np.nan,inplace=True)
df.replace('nan--',np.nan,inplace=True)
df.replace('Na--',np.nan,inplace=True)
```

```python
df["Last_Name"].replace(np.nan," ",inplace=True)
```

```python
df
```

```python
#Splitting address columns
df[["Street Address","State"]]=df["Address"].str.split(",",expand=True)
```

```python
df.replace("None",np.nan,inplace=True)
df
```

```python
df
```

```python
df.sort_values(by="First_Name",inplace=True)
```

```python
df.drop(columns=["Address"],inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns=["index"],inplace=True)
```

```python
df
```

```python
df.dropna(subset="Phone_Number",inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns=["index"],inplace=True)
```

```python
df["State"].replace("None","",inplace=True)
```
